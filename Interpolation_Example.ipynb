##### Librerias

```python
from scipy.interpolate import lagrange, InterpolatedUnivariateSpline
import numpy as np
import matplotlib.pyplot as plt
```

##### Datos

```python
x=[0,1,3,6]
y=[-3,0,5,7]
```

##### Graficando los datos

```python
plt.figure(figsize=(16,10))
plt.plot(x,y,'x', mew=3, label='Data')
plt.plot(x,y,'--', mew=3, label='Data')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='upper left', fontsize=14)
plt.show()
```

##### Interpolación por lagrange

```python
pol=lagrange(x,y)
```

#### Se obtiene el polinomio para unos puntos dados, ahora no tendremos 4 puntos si no 100


```python
xp=np.linspace(0,6,100)
yp=pol(xp)
```

#### Graficando la interpolación por Lagrange

```python
plt.figure(figsize=(16,10))
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(xp,yp,'X',mew=1, label='Interpolation',color='blue', alpha=0.2)
plt.plot(x,y,'--', mew=3, label='Linear', color='black')


plt.plot(x,y,'o', mew=6, label='Data', color='orange')

plt.legend(fontsize=14)
plt.show()
```

```python
plt.plot(x,y,'*', mew=2, label='Data')
plt.plot(xp,yp, label='Interpolation')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

#### Interpolación por splines

```python

spl_lineal=InterpolatedUnivariateSpline(x,y,k=1)

spl_k2=InterpolatedUnivariateSpline(x,y,k=2)
spl_k3=InterpolatedUnivariateSpline(x,y,k=3)
```

#### Se obtiene el polinomio para unos puntos dados, ahora no tendremos 4 puntos si no 100

```python
xspl=np.linspace(0,6,100)
yspl_lineal=spl_lineal(xspl)
yspl_k2=spl_k2(xspl)
yspl_k3=spl_k3(xspl)
```

```python
#### Graficando la interpolación por Splines
```

```python
plt.figure(figsize=(16,10))
plt.plot(x,y,'o', mew=5, label='Data', color='orange')
plt.plot(xspl,yspl_lineal, label='Spline k=1')
plt.plot(xspl,yspl_k2, label='Spline k=2')
plt.plot(xspl,yspl_k3, label='Spline k=3')
plt.plot(xp,yp, label='Spline k=2')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

##### Ahora comparemos los dos

```python
plt.plot(x,y,'*', mew=2, label='Data')
plt.plot(xp,yp, label='Interpolation Lagrange')

plt.plot(xspl,yspl, label='Interpolation Splines')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
